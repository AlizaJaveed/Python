### Sort, ORDER is imp (it matters), {permutations}



```python
import math

distinct_count = 5 
passcode_length = 6

def main():
    repeating_count = 2 
    total_outcomes = distinct_count * (math.factorial(passcode_length) // (math.factorial(repeating_count)))
    
    print('*** No of unique 6-digit passcodes is =', total_outcomes)
    print('*********************************')

if __name__ == '__main__':
    main()
```
